```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("C:/Users/user/Downloads/Dataset.csv")
df
```

```python
df.info()
df.describe()
df.isnull().sum()  #Check for missing values
```

```python
df_clean = df.dropna(axis=1, how='all')

#Check remaining columns
print(df_clean.columns)
```

```python
missing = df_clean.isnull().sum()
missing_columns = missing[missing > 0]
print(missing_columns)
```

```python
df[df.select_dtypes(include='number').columns] = df.select_dtypes(include='number').fillna(0)
df.head()
```

```python
df.isnull().sum()
```

```python
df.columns = [col.title() for col in df.columns]
df
```

```python
print(df['Country '].unique())
```

```python
df["Date "] = pd.to_datetime(df["Date "])
df["Year"] = df["Date "].dt.year
df["Month"] = df["Date "].dt.month
df["Day"] = df["Date "].dt.day

df
```

```python
df = df[~((df[' New Cases'].fillna(0) == 0) &
          (df['New_Deaths'].fillna(0) == 0) &
          (df['Vaccinated '].astype(str).str.lower() == 'unknown'))]
df
```

```python
# Replace 'Unknown' with 0
df = df.copy()
df['Vaccinated '] = df['Vaccinated '].replace('unknown', 0)

# Convert the column to numeric
df['Vaccinated '] = pd.to_numeric(df['Vaccinated '], errors='coerce')
df
```

```python
df.info()
df.describe(include='all')
```

```python
# Total number of cases and death
df[' New Cases'].sum()
df['New_Deaths'].sum()
```

```python
print(df.columns.tolist())
```

```python
# Country that has the highest number of cases
df[['Country ', ' New Cases']].sort_values(by=' New Cases', ascending=False)
```

```python
# Case Fatality Rate (CPR)
df = df.copy()
df['CPR (%)'] = (df['New_Deaths'] / df[' New Cases']) * 100
df
```

```python
# Total cases per country
df.groupby('Country ')[' New Cases'].mean().sort_values(ascending=False)
```

```python
df.set_index('Date ')[[' New Cases', 'New_Deaths', 'Vaccinated ']].plot(figsize=(10, 6), linewidth=2)
plt.title('Daily Trends in COVID-19 Cases, Deaths, and Vaccinated (November 2024)')
plt.xlabel('Date')
plt.ylabel('Count')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
country_summary = df.groupby('Country ')[[' New Cases', 'New_Deaths', 'Vaccinated ']].sum()
countries = country_summary.sort_values(by=' New Cases', ascending=False).head(10)

# Plot
countries.plot(kind='bar', figsize=(12,6), color=['green', 'red', 'yellow'])
plt.title('Country: Cases, Deaths, and Vaccinated')
plt.ylabel('count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df[' New Cases'], df['Vaccinated '], alpha=0.6, color='purple')
plt.title('Infected vs Vaccinated')
plt.xlabel(' New Cases')
plt.ylabel('Vaccinated ')
plt.grid(True)
plt.tight_layout()
plt.show()
```
